```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris_dataset=load_iris()

iris_df = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
iris_df.info()
iris_df.head()
```

```python
X=iris_dataset.data
y=iris_dataset.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=12)
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
accuracy=np.mean(y_pred==y_test)
print("Accuracy:",accuracy)

for i in range(len(X_test)):
    x = X_test[i]
    x_new = np.array([x])
    prediction = knn.predict(x_new)
    print("\n Actual : {0} {1}, Predicted :{2}{3}".format(y_test[i],iris_dataset["target_names"][y_test[i]],prediction,iris_dataset["target_names"][prediction]))

```

```python
import numpy as np
from collections import Counter

# Step 1: Generate 100 random values in [0, 1]
np.random.seed(0)  # For reproducibility
X = np.random.rand(100)

# Step 2: Label the first 50 points
y = []
for xi in X[:50]:
    if xi <= 0.5:
        y.append('C1')
    else:
        y.append('C2')

# Function to compute Euclidean distance (1D)
def euclidean_distance(x1, x2):
    return abs(x1 - x2)

# Step 3: KNN function
def knn_classify(x_new, X_train, y_train, k):
    # Compute distances from x_new to all training samples
    distances = [(xi, yi, euclidean_distance(x_new, xi)) for xi, yi in zip(X_train, y_train)]
    # Sort by distance
    distances.sort(key=lambda tup: tup[2])
    # Take k nearest neighbors
    k_neighbors = distances[:k]
    # Count the classes
    k_classes = [label for _, label, _ in k_neighbors]
    majority_class = Counter(k_classes).most_common(1)[0][0]
    return majority_class

# Step 4: Classify the remaining 50 points for multiple values of k
k_values = [1, 2, 3, 4, 5, 20, 30]
X_test = X[50:]  # Points to classify

# Step 5: Perform classification for each k
for k in k_values:
    print(f"\nResults for k = {k}")
    predictions = []
    for x_new in X_test:
        pred_class = knn_classify(x_new, X[:50], y, k)
        predictions.append(pred_class)
    # Display classification
    for i, (x_val, label) in enumerate(zip(X_test, predictions), start=51):
        print(f"x{i} = {x_val:.3f} => Classified as {label}")
```
